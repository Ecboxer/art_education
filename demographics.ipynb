Create a dataframe for each year and then merge them on dbn

```python
def clean_year(x): #Take the first year from the school year column
    return x[:4]
df['year'] = df['year'].apply(lambda x: clean_year(x))
```

```python
years = df['year'].unique()
init_cols = {i: i + '_2013' for i in df.columns if i != 'dbn'}
for i,y in enumerate(years):
    if i == 0:
        df_merged = df[df['year'] == y]
    else:
        df_merged = pd.merge(df_merged, df[df['year'] == y], how='outer', on='dbn', suffixes=['', '_{}'.format(years[i])])
df_merged.rename(index=str, columns=init_cols, inplace=True)
df_merged.rename(index=str, columns={'dbn': 'Q0_DBN'}, inplace=True)
#To match the art program dataset
df_merged.head()
```

Write to csv.

```python
df_merged.to_csv('resources/2013_18_Demographics_Wide.csv', index=False)
```
