```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('resources/2017_2018_Arts_Survey_data.csv')
df_demo = pd.read_csv('resources/2013_18_Demographics_Wide.csv')
```

Merge the arts survey and demographics data.

```python
df = df.merge(df_demo, on='Q0_DBN', how='left')
```

Merge the performance data.

```python
df_ela = pd.read_csv('resources/school-ela-results-2013-18-ALL.csv', skiprows=7)
df_math = pd.read_csv('resources/school-math-results-2013-18-ALL.csv', skiprows=7)
```

Rename columns.

```python
new_cols = ['sort', 'Q0_DBN', 'school_name', 'grade', 'year', 'category', 'num_tested', 'mean_score', 'num_1', 'perc_1', 'num_2', 'perc_2', 'num_3', 'perc_3', 'num_4', 'perc_4', 'num_34', 'perc_34']
cols_map = {df_ela.columns[i]: new_cols[i] for i in range(len(df_ela.columns))}
df_ela.rename(index=str, columns=cols_map, inplace=True)
df_math.rename(index=str, columns=cols_map, inplace=True)
```

```python
grades = df_ela['grade'].unique()
years = df_ela['year'].unique()
init_cols = {i: i + '_3_2013_ela' for i in df_ela.columns if i != 'Q0_DBN'}
for i,g in enumerate(grades):
    for j,y in enumerate(years):
        if i == j == 0:
            temp = df_ela[df_ela['year'] == y]
            df_merged = temp[temp['grade'] == g]
        else:
            temp = df_ela[df_ela['year'] == y]
            df_merged = pd.merge(df_merged, temp[temp['grade'] == g], how='outer', on='Q0_DBN', suffixes=['', '_{}_{}_ela'.format(grades[i], years[j])])
df_merged.rename(index=str, columns=init_cols, inplace=True)
df_merged.head()
```

```python
grades = df_math['grade'].unique()
years = df_math['year'].unique()
init_cols = {i: i + '_3_2013_math' for i in df_math.columns if i != 'Q0_DBN'}
for i,g in enumerate(grades):
    for j,y in enumerate(years):
        if i == j == 0:
            temp = df_math[df_math['year'] == y]
            df_merged_m = temp[temp['grade'] == g]
        else:
            temp = df_math[df_math['year'] == y]
            df_merged_m = pd.merge(df_merged_m, temp[temp['grade'] == g], how='outer', on='Q0_DBN', suffixes=['', '_{}_{}_math'.format(grades[i], years[j])])
df_merged_m.rename(index=str, columns=init_cols, inplace=True)
df_merged_m.head()
```

Merge ELA and math performance data with the arts and demographics data.

```python
df = df.merge(df_merged, on='Q0_DBN', how='left')
df = df.merge(df_merged_m, on='Q0_DBN', how='left')
df.head()
```
