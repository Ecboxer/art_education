```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('resources/school-ela-results-2013-18-ALL.csv', skiprows=7)
df.head()
```

Rename the columns for easier access.

```python
new_cols = ['sort', 'dbn', 'school_name', 'grade', 'year', 'category', 'num_tested', 'mean_score', 'num_1', 'perc_1', 'num_2', 'perc_2', 'num_3', 'perc_3', 'num_4', 'perc_4', 'num_34', 'perc_34']
cols_map = {df.columns[i]: new_cols[i] for i in range(len(df.columns))}
df.rename(index=str, columns=cols_map, inplace=True)
df.head()
```

No missing data

```python
df.shape[0] - df.count()
```

***

Create a dataframe for each year and grade, and merge them on dbn.

```python
grades = df['grade'].unique()
years = df['year'].unique()
init_cols = {i: i + '_3_2013' for i in df.columns if i != 'dbn'}
for i,g in enumerate(grades):
    for j,y in enumerate(years):
        if i == j == 0:
            temp = df[df['year'] == y]
            df_merged = temp[temp['grade'] == g]
        else:
            temp = df[df['year'] == y]
            df_merged = pd.merge(df_merged, temp[temp['grade'] == g], how='outer', on='dbn', suffixes=['', '_{}_{}'.format(grades[i], years[j])])
df_merged.rename(index=str, columns=init_cols, inplace=True)
df_merged.rename(index=str, columns={'dbn': 'Q0_DBN'}, inplace=True) #To match art program dataset
df_merged.head()
```

Import 17-18 school arts program data.

```python
df_art = pd.read_csv('resources/2017_2018_Arts_Survey_data.csv')
df_art.head()
```

```python
df_art_perf = df_art.merge(df_merged, on='Q0_DBN', how='left', sort=False)
#Left join for now and we will see how many, if any schools are without performance data
```

```python
df_art_perf.head()
```

Write csv of merged data.

```python
df_art_perf.to_csv('resources/17_18_Arts_and_Performance.csv', index=False)
```
